```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文显示（可选，避免图表乱码）
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 加载数据
df = pd.read_csv('./raw/housing.csv')

# 查看数据基本信息
print("数据形状:", df.shape)
print("\n前5行:")
print(df.head())
print("\n列信息:")
df.info()
print("\n缺失值统计:")
print(df.isnull().sum())
```

```python
# 精确查看每一列的缺失值数量
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 设置图表风格
sns.set_style("whitegrid")

# 1. 箱线图：用来发现“房价”和“收入”中的异常值
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['median_house_value'])
plt.title('房价分布箱线图 (可发现异常高价)')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['median_income'])
plt.title('收入分布箱线图')
plt.tight_layout()
plt.show()

# 2. 相关性热力图：看看哪些因素对房价影响最大
plt.figure(figsize=(10, 8))
# 只选择数值列来计算相关性，排除 ocean_proximity
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('特征相关性热力图')
plt.show()

# 3. 房价分布直方图：看房价是集中在某个区间，还是均匀分布
plt.figure(figsize=(10, 6))
plt.hist(df['median_house_value'], bins=50, edgecolor='black')
plt.title('房屋价值分布直方图')
plt.xlabel('房屋中位价值')
plt.ylabel('频次')
plt.show()
```

```python
# 计算中位数（使用中位数来填补空缺算是一种比较稳妥的方法，减少误差）
median_bedrooms = df['total_bedrooms'].median()

# 【修改点】使用 loc 或直接赋值的方式，避免链式赋值
df.loc[:, 'total_bedrooms'] = df['total_bedrooms'].fillna(median_bedrooms)

# 再次验证，现在缺失值应该为 0 了
print(df.isnull().sum())
```

```python
# 使用 pandas 的 get_dummies 进行 One-Hot 编码
# drop_first=True 可以避免虚拟变量陷阱，通常建议加上
df_encoded = pd.get_dummies(df, columns=['ocean_proximity'], drop_first=True)

# 查看编码后的新列
print(df_encoded.columns)
print(df_encoded.head())
```

```python
# 1. 创建新特征 (基于你清洗后的数据 df)
df['rooms_per_household'] = df['total_rooms'] / df['households']
df['bedrooms_per_room'] = df['total_bedrooms'] / df['total_rooms']
df['population_per_household'] = df['population'] / df['households']

# 2. (可选) 如果你还想使用 One-Hot 编码后的数据，可以基于 df_encoded 创建同样的特征
df_encoded['rooms_per_household'] = df_encoded['total_rooms'] / df_encoded['households']
df_encoded['bedrooms_per_room'] = df_encoded['total_bedrooms'] / df_encoded['total_rooms']
df_encoded['population_per_household'] = df_encoded['population'] / df_encoded['households']

# 3. 再检查一眼数据，应该能看到新加的三列了
print("新特征已添加，数据预览：")
print(df.head())
```

```python
# 导出清洗后的完整数据集
df.to_csv('./processed/housing_processed.csv', index=False)
print("数据集已保存为 housing_processed.csv，共 {} 条记录，{} 个特征。".format(df.shape[0], df.shape[1]))

# 如果你想同时保存 One-Hot 编码后的版本（供建模组员直接使用）
df_encoded.to_csv('./processed/housing_encoded.csv', index=False)
print("One-Hot编码版本已保存为 housing_encoded.csv，共 {} 条记录，{} 个特征。".format(df_encoded.shape[0], df_encoded.shape[1]))
```

## 数据清洗与特征工程总结

### 已完成的工作：
1. **缺失值处理**：使用中位数填充了 `total_bedrooms` 列的 207 个缺失值
2. **异常值分析**：通过箱线图发现房价和收入存在合理的高值区域，暂不做剔除
3. **特征编码**：对 `ocean_proximity` 进行了 One-Hot 编码，生成 4 个虚拟变量
4. **特征创造**：新增 3 个衍生特征
   - `rooms_per_household`：平均每户房间数
   - `bedrooms_per_room`：卧室占比
   - `population_per_household`：平均每户人口数

### 输出文件：
- `housing_processed.csv`：清洗后的完整数据集（共 20640 条记录，13 个特征）
